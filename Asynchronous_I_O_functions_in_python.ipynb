```python

# Asynchronous I/O function
# asyncio is a library to write concurrent code using async/await synta

import asyncio
import time

async def fetch_data(task_name, delay):
    print(f"{task_name} started")
# wait for seconds in delay
    await asyncio.sleep(delay)

    print(f"{task_name} completed after {delay} seconds")
    return f"{task_name} result"


async def main():
    start_time = time.perf_counter()

    results = await asyncio.gather(
        fetch_data("Task 1", 5),
        fetch_data("Task 2", 3),
        fetch_data("Task 3", 7)
    )

    end_time = time.perf_counter()

    print(results)
    print(f"Achieved in {end_time - start_time:.2f} seconds")


await main()
```

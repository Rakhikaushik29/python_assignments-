```python
 # Create a Student class and print details.
class Student:
 def __init__(self,name,age,marks,grade):
  self.name = name
  self.age = age
  self.marks = marks
  self.grade = grade
 def show_details(self):
    print(f"My name is {self.name} and i am {self.age} years old , and i got {self.marks} percent in my graduation and {self.grade} is my overall  grade")
```

```python
gunnu = Student("Gunnu",21,90,"A")
gunnu.show_details()
```

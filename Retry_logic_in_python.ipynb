```python
# Retry logic in python
# if we encounter Timeout , Connection error , HTTP error, retry the request
# Parameters: URL, max_retries , delay
# Function: response.raise_for_status()      ('errors come like 404,500 etc.')
# Default values in parameters

import requests
import time
url = "http://thetalentgrid.in/rakhi.php"

def fetch_url_with_retry(url, max_retries=5 ,delay=1):
  for attempt in range(1, max_retries + 1):
    print(f"Attempt no {attempt} to connect {url}")
    try:
      response = requests.get(url)
      response.raise_for_status()   # raises exception in case of 400 or 500
      return response
    except requests.exceptions.Timeout as e:
      print(f"Timeout error:{e}")
    except requests.exceptions.ConnectionError as e:
      print(f"Connection Error: {e}")
    except requests.exceptions.HTTPError as e:
      print(f"HTTP Error: {e}")

      if attempt < max_retries:
        print(f"Retrying in {delay} seconds...")
        time.sleep(delay)
  return None
```

```python
result = fetch_url_with_retry(url , 5 ,2)
```
